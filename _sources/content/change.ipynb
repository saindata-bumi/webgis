# Pemantauan perubahan setiap waktu menggunakan plot filmstrip citra satelit

## Latar Belakang

Memahami bagaimana bentangan alam dan manusia berubah seiring waktu untuk memberikan informasi penting tentang kesehatan ekosistem lokal dan perkembangan lingkungan binaan. Misalnya, data tentang perubahan dalam distribusi vegetasi di tempat tertent dapat digunakan untuk memantau dampak penggundulan hutan, atau melacak pemulihan hutan setelah pemulihan habitat atau peristiwa  kebakaran hutan. Pelacakan perubahan dalam wilayah perkotaan dapat digunakan untuk memantau pertumbuhan infrastruktur seperti pelabuhan dan jaringan transportasi, untuk kasus lain tentang perubahan pesisir dapat menjadi hal penting untuk memprediksi dan mengelola dampak erosi pantai atau hilangnya lahan  pesisir (misalnya hutan bakau).
